```python
# Import pandas
import pandas as pd

# Read in CSV
df = pd.read_csv('LoanStats_2018Q4.csv', header=1, skipfooter=2)
print(df.shape)
df.head()
```

```python
# Drop the 100% blank columns

df = df.drop(['url', 'member_id', 'desc', 'id'], axis=1)
```

```python
# Function to strip 'months' and whitespace from terms column entries

def term_to_int(contents):
  return int(contents.strip('months').strip())
```

```python
# Apply changes to term column to leave only integer values
df['term'] = df['term'].apply(term_to_int)

df['term'].value_counts()
```

```python
df['loan_status'].value_counts()
```

```python
# Creating a list to test function
statuses = ['Current', 'Fully Paid', 'Charged Off', 'In Grace']
```

```python
# Function to code loan status as 0's and 1's
# Current and Fully Paid get 1's while everything else gets 0's

def loan_state(status):
  if status == 'Current' or status == 'Fully Paid':
    return 1
  else:
    return 0

for status in statuses:
  print(loan_state(status))
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(loan_state)

df['loan_status_is_great'].value_counts()
```

```python
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)

df['last_pymnt_d'].dtypes
```

From what I can see online the <M8 data type is just like the datetime64 type but I'm getting floats back instead of ints after getting the month and year.

```python
df['last_pymnt_d'].head()
```

```python
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year

df.head()
```

```python
df['last_pymnt_d_month'].value_counts(dropna=False)
```

HA! They're floats because there are NaN values!

```python
df['last_pymnt_d_month'].fillna(0, inplace=True)
df['last_pymnt_d_year'].fillna(0, inplace=True)
```

```python
df['last_pymnt_d_month'].value_counts(dropna=False)
```

```python
df['last_pymnt_d_year'].value_counts(dropna=False)
```

```python
# Function to turn floats into ints

def float_to_int(value):
    return int(value)
```

```python
df['last_pymnt_d_month'] = df['last_pymnt_d_month'].apply(float_to_int)
df['last_pymnt_d_year'] = df['last_pymnt_d_year'].apply(float_to_int)
```
